```python
import numpy as np

# Charger le dataset
df = pd.read_csv('C:/Users/youss/OneDrive/Bureau/muse_v3.csv')

# Obtenir les combinaisons uniques de la colonne seeds
unique_seeds = df['seeds'].unique()

# Créer un dictionnaire qui associe chaque combinaison à la liste des musiques associées
comb_music_dict = {}
for seed in unique_seeds:
    # Sélectionner les lignes correspondant à la combinaison en question
    mask = df['seeds'] == seed
    seed_music = df[mask]['track'].tolist()
    seed_urls = df[mask]['lastfm_url'].tolist()
    # Ajouter la liste des musiques correspondantes au dictionnaire
    comb_music_dict[seed] = {'music': "+".join(seed_music), 'urls': "+".join(seed_urls)}

# Créer un nouveau DataFrame à partir du dictionnaire
new_df = pd.DataFrame.from_dict(comb_music_dict, orient='index')
new_df = new_df.reset_index()
new_df.columns = ['combination', 'music', 'urls']
```
